```python
import pandas as pd
import seaborn as sns

data = sns.load_dataset('titanic')
data.head()
```

```python
sns.displot(data['fare'])
```

```python
sns.displot(data['fare'], kde = True)
```

```python
sns.displot(data['fare'], bins = 10)
```

```python
sns.jointplot(x='age', y = 'fare', data=data)
```

```python
sns.jointplot(x='age', y='fare', data=data, kind = 'hex')
```

```python
data = data.dropna()
```

```python
sns.pairplot(data)
```

```python
sns.pairplot(data, hue='sex')
```

```python
sns.rugplot(data['fare'])
```
